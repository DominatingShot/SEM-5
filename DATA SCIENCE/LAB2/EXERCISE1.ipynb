```python
import pandas as pd
df = pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/autos/imports-85.data")
df.to_csv('lab2.csv')
```

```python
#add header
headers=["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","bodystyle","drive-wheels","engine-location","wheel-base","length","width","height","curbweight","engine-type","num-of-cylinders", "engine-size","fuelsystem","bore","stroke","compression-ratio","horsepower", 
"peak-rpm","city-mpg","highway-mpg","price"] 
df.columns=headers
df.columns
```

```python
df.head(5)
```

```python
import numpy as np
df.replace("?", np.nan, inplace = True)
df.head(5)

```

```python

print("missing values: ", df.isnull().sum().sum())
df.isnull().sum()
```

```python
l=df.isnull().sum()
l[l>0]
```

```python
print(df[df['normalized-losses'].isnull()].index.tolist()[:5])
print(len(df[df['normalized-losses'].isnull()].index.tolist()))
```

```python
df['normalized-losses'] = pd.to_numeric(df['normalized-losses'], errors='coerce')
df["stroke"] = pd.to_numeric(df["stroke"], errors="coerce")
df["bore"] = pd.to_numeric(df["bore"], errors="coerce")
df["normalized-losses"].fillna(df["normalized-losses"].mean(), inplace=True)
df["stroke"].fillna(df["stroke"].mean(), inplace=True)
df["bore"].fillna(df["bore"].mean(), inplace=True)
df.head(5)
df.isnull().sum()
```

```python
df["num-of-doors"]=pd.to_numeric(df["num-of-doors"], errors="coerce")
df["num-of-doors"].fillna(df["num-of-doors"].max(), inplace=True)
```

```python
#12. Replace the missing values in “horse_power”,peak_rpm with the values in the next row. 
#If the last row also contains missing values, replace them with the values in the previous row.
df["horsepower"]=pd.to_numeric(df["horsepower"], errors="coerce")
df["peak-rpm"]=pd.to_numeric(df["peak-rpm"], errors="coerce")
df["horsepower"].fillna(method='bfill', inplace=True)
df["peak-rpm"].fillna(method='bfill', inplace=True)
df.head(5)
```
