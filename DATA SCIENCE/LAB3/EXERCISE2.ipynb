```python
import missingno as msno 
import pandas as pd 
import numpy as np 
 # Create a sample DataFrame with missing values 
data = { 
'A': [1, 2, 3, np.nan, 5], 
'B': [6, np.nan, np.nan, 9, 10], 
'C': [11, 12, 13, 14, 15], 
} 
df = pd.DataFrame(data) 
# Create a matrix visualization of missing data 
msno.matrix(df)
```

```python
ti=pd.read_csv('titanic_dataset.csv')
msno.matrix(ti)#visualize the missing data
```

```python
msno.bar(ti)#visualize the missing data in bar chart
```

```python
ti2=ti.dropna()#drop the missing data
msno.matrix(ti2)#visualize the missing data
print(len(ti2),len(ti))#check the length of the data
```

```python
ti['Cabin'].fillna(ti['Cabin'].mode()[0], inplace=True)
print(ti['Cabin'].head(5))
```

```python
ti['Embarked'].fillna(ti['Embarked'].mode()[0], inplace=True)
print(ti['Embarked'].isnull().sum())
```

```python
#use knnimputer to fill in missing age
from sklearn.impute import KNNImputer
imputer = KNNImputer(n_neighbors=2)
ti['Age']=imputer.fit_transform(ti[['Age']])
print(ti['Age'].head(5))
```

```python
msno.bar(ti)#visualize the missing data
```

```python
ti3=pd.read_csv('titanic_dataset.csv')
#multivariate feauture imputation
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
imp = IterativeImputer()
ti3['Age']=imp.fit_transform(ti3[['Age']])
print(ti3['Age'].head(5))
msno.bar(ti3)#visualize the missing data
```
