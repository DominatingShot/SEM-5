```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('cupcake.csv') 
#read the data from the csv file
```

```python
mincup=df['Cupcake'].min()
maxcup=df['Cupcake'].max()
print(mincup,maxcup)
labels=["small","medium","big"]
bins=np.linspace(mincup,maxcup,4)
print(bins)
```

```python
cuts=pd.cut(df['Cupcake'],bins,labels=labels,include_lowest=True)
plt.hist(cuts)
```
